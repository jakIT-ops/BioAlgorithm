## 1.3 Биологийн үүрэгтэй молекулууд

- **Нуклейн хүчил (Nucleic acids):**
    - Эс дотор хадгалагдаж байдаг удамшлын кодыг агуулдаг
    - **Дезоксирибонуклеины хүчил (ДНХ)** ба **Рибонуклеины хүчил (РНХ).** Химийн хувьд маш төстэй.
    - ДНХ нь эсийг төрүүлж, үүргээ гүйцэтгэн ажиллаж байхад шаардлагатай мэдээллийг агуулдаг.
    - ДНХ нь эукариот эсүүдийн цөмд, прокариот эсүүдийн цитоплазмд байрлана.
    - ДНХ-д кодлогдсон мэдээлэл дэх найрлагаыг бүрдэх явцад РНХ нь дундын материалаар хэрэглэгддэг.


- **Амин хүчил (Amino acids):**

    - Био-молекулын өөр нэг чухал төрөл юм.
    - Эс дотор ихэнх үүргийг гүйцэтгэдэг макромолекулууд болох **уураг(proteins)** –г бүрдүүлэгч блокууд
        - Жнь: Энзим (Enzymes) - химийн урвалыг дэмждэг харьцангуй их тохиолддог төрөл. Зарим молекулыг эсийн үүргийн биелэлтэнд шаардлагатай өөр төрлийн молекулд хувиргадаг.
        
        
- Бусад чухал органик макромолекулууд:
    - **Нүүрс ус (Carbohydrates):** Шууд болон урт хугацаанд шаардлагатай энергийг хуримтлуулдаг.
    - **Липид (Lipids):** Плазмын мембраны бүрдэл хэсэг. Сигнал болон энергийн хуримтлалыг идэвхжүүлэх үүрэгтэй.

2. **Consider a word of length 5. Indicate the number of possible word combinations:**

```python
def ValidGeneticCode(seq):
  validDna = seq.count("A") + seq.count("C") + seq.count("G") + seq.count("T")
  validRna = seq.count("C") + seq.count("G") + seq.count("A") + seq.count("U")
  if(validDna == len(seq)):
    print("based on the DNA!") 
  elif(validRna == len(seq)):
    print("based on the RNA!")
  else:
    print("based on the Protein!")

sequence = input("5 - ын урттай оруулна уу: ").upper()
ValidGeneticCode(sequence)
```

```python
RNA_Codons = {
    "GCU": "Ala", "GCC": "Ala", "GCA": "Ala", "GCG": "Ala",
    "UGU": "Cys", "UGC": "Cys",
    "GAU": "Asp", "GAC": "Asp",
    "GAA": "Gin", "GAG": "Gin",
    "UUU": "Phe", "UUC": "Phe",
    "GGU": "Gly", "GGC": "Gly", "GGA": "Gly", "GGG": "Gly",
    "CAU": "His", "CAC": "His",
    "AUA": "lle", "AUU": "lle", "AUC": "lle",
    "AAA": "Lys", "AAG": "Lys",
    "UUA": "Leu", "UUG": "Leu", "CUU": "Leu", "CUC": "Leu", "CUA": "Leu", "CUG": "Leu",
    "AUG": "Met",
    "AAU": "Asn", "AAC": "Asn",
    "CCU": "Pro", "CCC": "Pro", "CCA": "Pro", "CCG": "Pro",
    "CAA": "Gin", "CAG": "Gin",
    "CGU": "Arg", "CGC": "Arg", "CGA": "Arg", "CGG": "Arg", "AGA": "Arg", "AGG": "Arg",
    "UCU": "Ser", "UCC": "Ser", "UCA": "Ser", "UCG": "Ser", "AGU": "Ser", "AGC": "Ser",
    "ACU": "Thr", "ACC": "Thr", "ACA": "Thr", "ACG": "Thr",
    "GUU": "Val", "GUC": "Val", "GUA": "Val", "GUG": "Val",
    "UGG": "Trp",
    "UAU": "Tyr", "UAC": "Tyr",
    "UAA": "Stop", "UAG": "Stop", "UGA": "Stop"
}
```

```python
class RNA():
  def __init__(self, s):
    self.seq = s.upper()

  def codons(self):
    # 3 3 аар нь таслана
    codon_lst = []
    for i in range(0, len(self.seq), 3):
      codon_lst.append(self.seq[i:i+3])
    if(len(codon_lst[-1])!=3):
      codon_lst.remove(codon_lst[-1])
    return codon_lst

  def count(self):
    # codons дунд байгаа эсэхийн шалгана
    codon_lst = self.codons()
    complete = 0
    incomplete = 0
    for c in codon_lst:
      if(c in codon):
        complete += 1
      else: 
        incomplete += 1
    return "байгаа " + str(complete), "байхгүй " + str(incomplete)
  
  def startEnd(self, s='AUG'):
    # Met-Stop
    codon_lst = list(self.codons())
    start = codon_lst.index(s)
    end = 0
    for i in range(start, len(codon_lst)):
      if(codon_lst[i] == 'UAA' or codon_lst[i] == 'UGA' or codon_lst[i] == 'UAG'):
        end = i
        break;
    return "Эхлэх цэг " + str(start), "Төгсөх цэг " + str(end)
  
  def readingFrame(self, init_pos = 1):
    #Phe - Leu
    res = []
    for pos in range(init_pos, len(self.seq) - 2, 3):
        res.append(RNA_Codons[self.seq[pos:pos + 3]])
    #return [RNA_Codons[self.seq[pos:pos + 3]] for pos in range(init_pos, len(self.seq) - 2, 3)]
    return res

  def rem(self):
     self.seq = self.codons()
     self.seq.remove(self.seq[-1])
     return self.seq
     # zasah

seqRna = "UUUUUCAUGUCCGGGAUGUAG"
rna = RNA(seqRna)
print(rna.codons())
#print(dna.rem())
print(rna.startEnd())
print(rna.count())
print(rna.readingFrame())
```

#5. Suppose that you are studying a very short protein sequence. We would like to know the sequence of nucleotides that gave rise to this protein sequence: Met-Ala-His-Trp.

a. How many mRNA sequences can code this protein?

```python
import itertools as it

seq = {'Met' : ['AUG'], 'Ala': ['GCU','GCC','GCA','GCG'],'His':['CAU', 'CAG'],'Trp': ['ACU','ACC','ACA','ACG']}
keys, values = zip(*seq.items()) 
permutations_dicts = [dict(zip(keys, v)) for v in it.product(*values)]
# zip Энэ нь аргумент болгон дамжуулсан бүх утгуудыг хос хэлбэрээр хадгалдаг
# dict() create dictionary
# product('ABCD', repeat=2) res = AA AB AC AD BA BB BC BD CA CB CC CD DA DB DC DD

print(permutations_dicts)
print("Rna seq: ", len(permutations_dicts))
print(permutations_dicts)
```

b. What is the length of the mRNA?

```python
print(len(permutations_dicts[0]['Met']) + len(permutations_dicts[0]['Ala'])+len(permutations_dicts[0]['His'])+len(permutations_dicts[0]['Trp']))
```
